# Analysis 4: Plot linear regressions of data in `data/` using Pandas and Seaborn

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
print(pd.__version__)
print(sns.__version__)
```

```python
indir = 'data/'

fname_1 = 'Sep2021_Liaoning_data.csv'
fname_2 = 'Sep2021_Jilin_data.csv'
fname_3 = 'Sep2021_Heilongjiang_data.csv'

provs = ["Liaoning", "Jilin", "Heilongjiang"]

yield_1 = pd.read_csv(f"{indir}/{fname_1}", index_col=0)
yield_2 = pd.read_csv(f"{indir}/{fname_2}", index_col=0)
yield_3 = pd.read_csv(f"{indir}/{fname_3}", index_col=0)
```

```python
yield_1.head()
```

```python
fig = plt.figure(figsize=(16, 12))

for i, data in enumerate([yield_1, yield_2, yield_3]):
    for j, x in enumerate(["Jun T", "Jul T", "Aug T"]):
        sns.regplot(x = "Jun T",
                    y = "Maize yield anomaly",
                    data = yield_1,
                    label = provs[i],
                    ax = fig.add_subplot(3,3,i*3+j+1),
                   ).set(title=provs[i])
        
plt.tight_layout()
```
